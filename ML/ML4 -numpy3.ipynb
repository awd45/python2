# creation function
## numpy에서 array를 만드는 function

```python
import numpy as np
```

```python
# array의 범위를 지정하여, 값의 list를 생성하는 명령어

np.arange(30) # range : list의 range와 같은 효과, integer로 0부터 29까지 배열 추출
```

```python
np.arange(0, 5, 0.5) # floating point도 표시가능함
# np.arange(시작, 끝, step)
```

```python
np.arange(0, 5, 0.5).tolist() # tolist형으로 변환가능 (list에서는 step을 floatingType으로는 출력x)
```

```python
np.arange(30).reshape(5,6)
```

```python
np.arange(30).reshape(3,5,2) # 3차원으로 표현가능
```

# ones, zeros and empty
## zeros '0'으로 가득찬 ndarray 생성

### np.zeros(shape, dtype, order)

```python
np.zeros(shape = (10,), dtype=np.int8) # 0으로 가득참
```

```python
np.zeros((2,5))
```

## ones '1'로 가득찬 ndarray 생성
### np.ones(shape,dtype,order)

```python
np.ones(shape=(10,), dtype=np.int8)
```

```python
np.ones((2,5))
```

## empty 'shape'만 주어지고 비어있는 ndarray 생성
## (memory initialization(메모리초기화)이 되지 않음)

```python
np.empty(shape=(10,), dtype=np.int8) # 실해할때마다 메모리에 남아있던값들이 튀어나옴
```

```python
np.empty((3,5))
```

# something like
## 기존 ndarray의 shape 크기 만큼 1,0 또는 empty array를 반환

```python
test_matrix = np.arange(30).reshape(5,6)
np.ones_like(test_matrix)
```

```python
np.zeros_like(test_matrix)
# 기존의 ndarray의 shape대로 0이나 1로 채워진 array를 받는다.
```

# identity 단위행렬만들기
## 단위행렬(i 행렬)을 생성함
### n -> number of rows

```python
np.identity(n=3, dtype=np.int8)
```

```python
np.identity(5)
```

# eye 
## 대각선이 1인 행렬, k값의 시작 index의 변경이 가능

```python
np.eye(N=3, M=5, dtype=np.int8)
```

```python
np.eye(3)
```

```python
np.eye(3,5,k=2)
# k는 start index
```

# random sampling
## 데이터분포에 따른 sampling으로 array를 생성

```python
np.random.uniform(0,1,10).reshape(2,5) # 균등분포
# np.random.uniform(최저값 , 최고값 ,데이터갯수)
```

```python
np.random.normal(0,1,10).reshape(2,5) # 정규분포
# np.random.normal(평균값, 값들이 벌어진정도, 거기서 10의 값)
```
