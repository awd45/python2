# axis (축)
## 모든 operation function을 실행할때, 기준이 되는 dimension축

```python
test_array = np.arange(1,13).reshape(3,4)
test_array
```

```python
test_array.sum(axis=1), test_array.sum(axis=0)
```

```python
test_array = np.arange(1,13).reshape(3,4)
test_array
```

```python
test_array.mean(), test_array.mean(axis=0)
```

```python
test_array.mean(), test_array.mean(axis=0)
```

```python
test_array.std(), test_array.std(axis=0)
```


## vertical stack 수직으로 어레이합치기
## horizontal 수평으로 어레이합치기

```python
a = np.array([1,2,3])
b = np.array([2,3,4])
np.vstack((a,b))
```

```python
a = np.array([[1],[2],[3]])
b = np.array([[2],[3],[4]])
np.hstack((a,b))
```
